```python
import time
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
def translate_english_to_french(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated_tokens = model.generate(**inputs)
    translated_text = tokenizer.decode(
        translated_tokens[0], skip_special_tokens=True
    )
    return translated_text
```

```python
def typing_effect(text, delay=0.05):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()
```

```python
print("🇫🇷 English → French Translation Chatbot")
print("Type 'exit' to quit\n")

while True:
    user_input = input("You (English): ")

    if user_input.lower() == "exit":
        print("👋 Chatbot exited. Goodbye!")
        break

    french_translation = translate_english_to_french(user_input)

    print("Bot (French): ", end="")
    typing_effect(french_translation)
```
