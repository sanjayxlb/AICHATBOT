```python
import time
from transformers import pipeline
# Import necessary modules for text generation and typing animation.
```

```python
poet = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)
# Load the FLAN-T5 instruction-tuned model to generate poetic responses.
```

```python
def typing_effect(text, line_delay=0.4, word_delay=0.08):
    print("PoetBot:")
    for line in text.split("\n"):
        print("  ", end="", flush=True)
        for word in line.split():
            print(word, end=" ", flush=True)
            time.sleep(word_delay)
        print()
        time.sleep(line_delay)
# This function controls the timed output of generated text using the time module.
# It iterates through each line and word to introduce controlled delays.
# Word-level and line-level pauses simulate incremental text rendering.
# This technique improves readability and mimics real-time response generation.
```

```python
def poetic_response(user_input):
    prompt = (
        "You are a poetic personality chatbot.\n"
        "You NEVER answer facts directly.\n"
        "You ALWAYS respond with a poem of exactly 4 lines.\n"
        "Do not repeat the same phrase.\n"
        "Use metaphor and emotion.\n"
        "Do not explain.\n\n"
        f"User input: {user_input}\n"
        "Poem:\n"
    )

    output = poet(
        prompt,
        max_new_tokens=80,
        repetition_penalty=1.8
    )

    poem = output[0]["generated_text"].strip()
    lines = poem.split("\n")[:4]
    return "\n".join(lines)
# This function builds a structured prompt that defines the chatbotâ€™s poetic behavior.
# The prompt is passed to the language model to generate a verse-based response.
# Generation parameters control length, repetition, and creativity of the output.
# The returned text is processed to ensure consistent poetic formatting.

```

```python
print(" Welcome to the Poetic Personality Chatbot ")
print("Type 'quit' to exit\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "quit":
        typing_effect("May your thoughts rest gently in quiet light.")
        break

    reply = poetic_response(user_input)
    typing_effect(reply)
# This loop continuously listens for user input during the chatbot session.
# It evaluates exit conditions and controls the flow of conversation.
# Each user input is processed and passed to the response generation function.
# The loop enables real-time, interactive communication with the chatbot.
```
